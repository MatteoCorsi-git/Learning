```r
# LEARNING TIDYVERSE
library(tidyverse)

# import a dataset
ds <- read_csv("msleep.csv")
```

```r
# explore with function glimpse()
glimpse(ds)

# Use filter() to subset msleep to only herbivores.
# glimpse %>% filter()
```

```r
# SUBSETTING ROWS WITH FILTER()

# Use filter() to subset msleep to only herbivores.

# Inspect the values of the variable "vore"
unique(ds$vore)

# Use filter() to subset msleep to only herbivores.
ds_herbi <- ds %>%
  filter(vore == "herbi")

head(ds_herbi)
```

```r
# Use filter() to subset msleep to only animals who are awake for at least 12 hours of the day.
ds_aw12 <- ds %>%
  filter(awake >= 12)

head(ds_aw12)
```

```r
# Use filter() to subset msleep to only herbivores who are awake for at least 12 hours of the days.
ds_herb12 <- ds %>%
  filter(vore == "herbi", awake >= 12)

head(ds_herb12)
```

```r
# Use filter() and logical operator %in% (or |) to subset the dataset to keep just herbivores and carnivores.

ds %>%
  filter(vore %in% c("herbi", "carni"))

# OR

ds %>%
  filter(vore == "herbi" | vore == "carni")
```

```r
# Use filter() to subset msleep to include only herbivores and carnivores who sleep at least 12 hours a day.

ds %>%
  filter (vore %in% c("herbi", "carni"),
                    sleep_total >= 12)

# OR

ds %>%
  filter((vore == "herbi" | vore == "carni"),
        sleep_total >= 12)
```

```r
# Use filter() to subset the data to only carnivores who weigh more than 50 kg.

ds %>%
  filter(vore == "carni", bodywt > 50)
```

```r
# Modify your code to the previous question so that it uses variables rather than values in filter()
# NB: this code demonstrate how much simpler is to modify the code when we use variables instead of values

target_vore <- "herbi"
target_weight <- 25

ds %>%
  filter(vore == target_vore, bodywt > target_weight)
```

```r
# SUBSETTING COLUMNS WITH SELECT()

# Use select() to keep only the columns name, awake, sleep_total, sleep_rem, and sleep_cycle.
ds %>%
  select(name, awake, sleep_total, sleep_rem, sleep_cycle)
```

```r
# Use select() to remove the columns genus and order.
ds %>%
  select(-order, -genus)
```

```r
# You can also use select() to re-order columns. Reorder the columns as: bodywt, brainwt, then everything else (everything())
ds %>%
  select(bodywt, brainwt, everything())
```

```r
# COUNTING THE NUMBER OF ROWS

# The simpler way is nrow(), even if we cannot pipe out another dataframe from it
nrow(ds)
# OR
ds %>%
  nrow()

# A nice Dplyr function is tally()
ds %>%
  tally()

# Answer the question: How many mammals weigh more than 2000 kg?
ds %>%
  filter(bodywt > 2000) %>%
  tally()
```

```r
# CREATING NEW COLUMNS WITH MUTATE()

# Use mutate() to create a new column called class which literally just contains the string â€œMammaliaâ€. Indeed, these are all mammals!
# Play with it...

ds %>%
  mutate(class = "Mammalia") %>% # add the column
  select(class, everything()) %>% # order the df to check the correctness of the new column
  filter(vore == "herbi") %>%
  tally()
```

```r
# Practice writing pipeline code one line at a time below. For each step of the pipeline, look at the result to confirm it worked before adding the next step.:
# First, use mutate() to add a new column to msleep called class that contains the string â€œMammaliaâ€. Look at the output to make sure it worked.
# Then, add onto the pipeline: code with filter() to REMOVE all mammals in the order â€œRodentiaâ€ (hint: remember the !=logical operator!).
# Finally, add onto the pipeline: use select() to only keep columns in this order: class, order, genus, name.
# If all done correctly, the result should have 4 columns and 61 rows.

ds %>%
  mutate(class = "Mammalia") %>%
  filter(order != "Rodentia") %>%
  select(class, order, genus, name)
```

```r
# Create a new variable called bodywt_lbs which contains the body weight in pounds (1 kg = 2.2 lbs).

ds %>%
  mutate(bodywt_lbs = bodywt * 2.2) %>%
  select(bodywt, bodywt_lbs)
```

```r
glimpse(ds)
```

```r
# Use mutate() to create a new column called percent_day_awake that gives the percentage of the day that each species spends awake.

ds %>%
  mutate(percent_day_awake = (awake / 24) * 100) %>%
  select(percent_day_awake)
```

```r
# Use mutate() to create a new column called log_bodywt that gives the natural logarithm of the body weight.

ds %>%
  mutate(log_bodywt = log(bodywt)) %>%
  select(bodywt, log_bodywt)
```

```r
# Use mutate() to create a new column called sleep_awake_ratio that has the ratio of total time spent asleep to total time spent awake (sleep_total divided by awake)

ds %>%
  mutate(sleep_awake_ratio = sleep_total / awake) %>%
  select(sleep_total, awake, sleep_awake_ratio)
```

```r
# Letâ€™s take a quick detour out of dplyr and into the package tidyr, which is part of the tidyverse.
# This package (which has been loaded for you!) contains a function drop_na() which removes NAâ€™s from a tibble.
# Explore the use of this function with get_help("drop_na"), and then use the function to remove all rows from msleep that contain NAs in the following columns:

?drop_na() # It drops the rows that have at least one missing value

# check the number of missing values
sum(is.na(ds$brainwt))
sum(is.na(ds$sleep_cycle))

ds %>%
  select(brainwt, sleep_cycle) %>%
  drop_na()

# OR (faster)

ds %>%
  drop_na(brainwt, sleep_total)
```

```r
# add a another pipe that creates TWO new columns with mutate() containing the mean brainwt and then median bodywt for all mammals.
# Name these columns mean_brainwt and median_bodywt

ds %>%
  mutate(mean_brainwt = round(mean(brainwt, na.rm = TRUE), 2), median_bodywt = median(bodywt)) %>%
  select(mean_brainwt, median_bodywt)
```
