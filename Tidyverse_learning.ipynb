```r
# add a another pipe that creates TWO new columns with mutate() containing the mean brainwt and then median bodywt for all mammals.
# Name these columns mean_brainwt and median_bodywt
ds %>%
  mutate(mean_brainwt = mean(brainwt, na.rm = TRUE),
         median_bodywt = median(bodywt, na.rm = TRUE)) %>%
  select(mean_brainwt, median_bodywt)

# OR

ds %>%
  drop_na(brainwt, bodywt) %>%
  mutate(mean_brainwt = mean(brainwt),
         median_bodywt = median(bodywt)) %>%
  select(mean_brainwt, median_bodywt)
```

```r
# Use the if_else() function to create another column called are_you_herbi.
# This column should contain the value â€œherbivoreâ€ if yes, and â€œnot_an_herbivoreâ€ if no.
ds %>%
  mutate(are_you_herbi = if_else(vore == "herbi", "herbivore", "not_an_herbi")) %>%
  select(vore, are_you_herbi)
```

```r
# Create a column called â€œweight_classâ€, where mammals that greater than or equal to 100 kg are considered â€œheavyâ€ and mammals that weigh less than 100 kg are considered â€œlightâ€

ds %>%
  mutate(weight_class = if_else(bodywt >= 100, "heavy", "light")) %>%
  select(bodywt, weight_class)
```

```r
# Create a new column in msleep called needs_more_caffeine where mammals who are awake more than 16 hours a day have the value â€œdefinitelyâ€ and other mammals have â€œnopeâ€

ds %>%
  mutate(needs_more_caffeine = if_else(awake > 16, "definitely", "nope")) %>%
  select(awake, needs_more_caffeine)
```
